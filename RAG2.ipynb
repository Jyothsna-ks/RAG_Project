```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("rag_customer_support_faq.pdf")

docs = loader.load()

print(len(docs))
```

```python
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50
)

chunks = splitter.split_documents(docs)

print(len(chunks))
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings()
```

```python
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(chunks, embedding)

print("Database created")
```

```python
query = "What is refund policy?"

results = db.similarity_search(query, k=1)

print(results[0].page_content)
```

```python
def answer_question(query):
    results = db.similarity_search(query, k=1)
    best_result = results[0].page_content
    print(f"Q: {query}")
    print(f"A: {best_result}")
    print("-" * 40)

# Test it
answer_question("What is refund policy?")
answer_question("How do I track my order?")
answer_question("How to contact support?")
```

```python
def smart_answer(query, min_length=50):
    results = db.similarity_search(query, k=1)
    answer = results[0].page_content

    if len(answer) < min_length:
        print(f"Q: {query}")
        print("⚠️  Low confidence — Escalating to human support agent...")
    else:
        print(f"Q: {query}")
        print(f"✅ Answer: {answer}")
    print("-" * 40)

# Test
smart_answer("What is refund policy?")
smart_answer("xyz")
```
